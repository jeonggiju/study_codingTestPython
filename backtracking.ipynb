```python
import sys
input = sys.stdin.readline

def f(k):
    if k == m:
        print(' '.join(map(str, nums)))
        return
    
    for i in range(1, n + 1):
        if not is_used[i]:
            nums[i] = i
            is_used[i] = True
            f(k + 1)
            is_used[i] = False

if __name__ == "__main__":
    n, m = map(int, input().split())
    nums = [0] * m
    is_used = [False] * (n + 1) # 1-indexed
    f(0)
```

```python
import sys
input = sys.stdin.readline

def f(cur):
    if cur == n:
        global count
        count += 1
        return

    for i in range(n):
        if not is_used_horiz[i] and not is_used_diag1[cur + i] and not is_used_diag2[cur - i + n - 1]:
            is_used_horiz[i] = is_used_diag1[cur + i] = is_used_diag2[cur - i + n - 1] = True
            f(cur + 1)
            is_used_horiz[i] = is_used_diag1[cur + i] = is_used_diag2[cur - i + n - 1] = False

if __name__ == "__main__":
    n = int(input())
    is_used_horiz = [False] * n
    is_used_diag1 = [False] * (2 * n - 1)
    is_used_diag2 = [False] * (2 * n - 1)
    count = 0
    
    f(0)
    print(count)
```

```python
import sys
input = sys.stdin.readline

def f(cur, tot):
    if cur == n:
        global count
        if tot == s:
            count += 1
        return

    f(cur + 1, tot)
    f(cur + 1, tot + nums[cur])

if __name__ == "__main__":
    n, s = map(int, input().split())
    nums = list(map(int, input().split()))
    count = 0
    f(0, 0)
    if s == 0:
        count -= 1
    print(count)
```

### [1. N과 M (2) (15650)](https://www.acmicpc.net/problem/15650)

#### 1st Trial

```python
import sys
input = sys.stdin.readline

def f(k):
    if k == m:
        print(" ".join(map(str, nums)))
        return
    
    for i in range(1, n+1):
        if not is_used[i]:
            if k == 0 or i > nums[k-1]:
                nums[k] = i
                is_used[i] = True
                f(k+1)
                is_used[i] = False

if __name__ == "__main__":
    n, m = map(int, input().split())
    nums = [0] * m
    is_used = [False] * (n+1)
    f(0)
```

#### 2nd Solution

- Improve efficiency by starting the next loop from the last selected number + 1 instead of always starting from 1. This would avoid unnecessary checks for numbers less than `a[k - 1]`.
- Remove the `is_used` by leveraging the `start` parameter to ensure that each subsequent number is greater than the previous one, thus maintaining ascending order and avoiding duplicates. This approach inherently ensures that each number is used at most once in each combination without the need for explicitly tracking used numbers with an additional array.

```python
import sys
input = sys.stdin.readline

def recur(cur, start):
    if cur == m:
        print(" ".join(map(str, nums)))
        return
    
    for i in range(start, n + 1):
        nums[cur] = i
        recur(cur + 1, i + 1)

if __name__ == "__main__":
    n, m = map(int, input().split())
    nums = [0] * m
    recur(0, 1)
```

### [4. 연산자 끼워넣기 (14888)](https://www.acmicpc.net/problem/14888)

#### 1st Trial


```python
import sys
input = sys.stdin.readline

def calculate(left, operator, right):
    if operator == 0:
        return left + right
    elif operator == 1:
        return left - right
    elif operator == 2:
        return left * right
    elif operator == 3:
        if left < 0:
            return -(-left // right)
        else:
            return left // right
    raise ValueError("Invalid operator")

def recur(cur, left):
    if cur == n - 1:
        res.append(left)
        return
    
    for i in range(4):
        if operators[i] > 0:
            operators[i] -= 1
            recur(cur + 1, calculate(left, i, nums[cur + 1]))
            operators[i] += 1


if __name__ == "__main__":
    n = int(input())
    nums = list(map(int, input().split()))
    operators = list(map(int, input().split()))
    res = []
    
    recur(0, nums[0])
    res.sort()
    print(res[-1])
    print(res[0])
```

### [5. 부분수열의 합 (1182)](https://www.acmicpc.net/problem/1182)

```python
import sys
input = sys.stdin.readline

def recur(cur, tot):
    global count
    if cur == n:
        if tot == s:
            count += 1
        return
    
    recur(cur + 1, tot)
    recur(cur + 1, tot + nums[cur])

if __name__ == "__main__":
    n, s = map(int, input().split())
    nums = list(map(int, input().split()))
    count = 0
    recur(0, 0)
    if s == 0:
        count -= 1  # empty set
    print(count)
```

#### 2nd Trial

- Prune Infeasible Paths Early:<br>
    If at any point the toal weight (`tot`) of the visited cities exceeds `min_wieght`, the recursion can retrun early.
- Fix Starting City:<br>
    Since the problem is a cycle, you can fix the starting city to be 0 (for example) to avoid redundant permutations that are essentially the same tour starting at different cities.


```python
import sys
input = sys.stdin.readline
MAX = 10_000_000  # 1_000_000 * 10

def recur(cur, tot):
    global min_weight
    if tot >= min_weight:  # Pruning
        return
    if cur == n:
        if w[route[cur-1]][0] == 0:  # No path back to start
            return
        min_weight = min(min_weight, tot + w[route[cur-1]][0])
        return

    for city in range(1, n):  # Start from 1 since 0 is the starting city
        cur_weight = w[route[cur-1]][city]
        if not visited[city] and cur_weight != 0:
            route[cur] = city
            visited[city] = True
            recur(cur + 1, tot + cur_weight)
            visited[city] = False


if __name__ == "__main__":
    n = int(input().strip())
    w = [list(map(int, input().split())) for _ in range(n)]
    visited = [False] * n
    route = [-1] * (n + 1)
    route[0] = 0
    min_weight = MAX
    recur(1, 0)
    print(min_weight)
```

### \* [외판원 순회 (2098)](https://www.acmicpc.net/problem/2098)

#### Solution by GPT-4

```python
import sys

MAX = sys.maxsize

def tsp(dp, w, mask, pos, n):
    # If all cities have been visited
    if mask == (1 << n) - 1:
        return w[pos][0] if w[pos][0] > 0 else MAX

    # Return already computed value
    if dp[mask][pos] != -1:
        return dp[mask][pos]

    min_cost = MAX
    for city in range(n):
        if mask & (1 << city) == 0 and w[pos][city] > 0:
            new_cost = w[pos][city] + tsp(dp, w, mask | (1 << city), city, n)
            min_cost = min(min_cost, new_cost)

    dp[mask][pos] = min_cost
    return min_cost

if __name__ == "__main__":
    n = int(input().strip())
    w = [list(map(int, input().split())) for _ in range(n)]

    # Initialize DP table
    dp = [[-1 for _ in range(n)] for _ in range(1 << n)]
    
    print(tsp(dp, w, 1, 0, n))  # Start from city 0 with only city 0 visited
```

- Proceed with the cell by cell instead of row by row.
- For each cell in the current row:
  - If the cell is empty, try placing each possible number.
  - Check if placing that number is valid.
  - If valid, place the number and recursively solve for the next cell/row.
  - If placing any number in this cell doesn't lead to a solution, reset the cell and backtrack.

This problem requires careful handling of the backtracking conditions and ensuring that you reset the state correctly when backtracking. Focus on how you're marking cells as visited and how you undo these markings when a particular path doesn't lead to a solution.

```python
import sys
input = sys.stdin.readline

def get_subgrid_index(row, col):
    return (row // 3) * 3 + (col // 3)
        
def recur(row, col):
    # If we've reached the 9th row, we've filled all cells correctly
    if row == 9:
        for i in range(9):
            print(" ".join(map(str, board[i])))
        return True
    
    # If we've reached the end of a row, move to the next row
    if col == 9:
        return recur(row + 1, 0)
    
    # If the current cell is already filled, move to the next cell
    if board[row][col] != 0:
        return recur(row, col + 1)
    
    grid = get_subgrid_index(row, col)
    for num in range(1, 10):
        if not is_used_col[col][num] and not is_used_row[row][num] and not is_used_subgrid[grid][num]:
            board[row][col] = num
            is_used_col[col][num] = is_used_row[row][num] = is_used_subgrid[grid][num] = True

            # Move to the next cell
            if recur(row, col + 1):
                return True
            
            # If placing num doesn't lead to a solution, backtrack
            board[row][col] = 0
            is_used_col[col][num] = is_used_row[row][num] = is_used_subgrid[grid][num] = False

    # If no number can be placed in the current cell, return False
    return False

if __name__ == "__main__":
    board = [list(map(int, input().split())) for _ in range(9)]
    is_used_col = [[False] * 10 for _ in range(9)]
    is_used_row = [[False] * 10 for _ in range(9)]
    is_used_subgrid = [[False] * 10 for _ in range(9)]
    for i in range(9):
        for j in range(9):
            if board[i][j] != 0:
                is_used_col[j][board[i][j]] = True
                is_used_row[i][board[i][j]] = True
                is_used_subgrid[get_subgrid_index(i, j)][board[i][j]] = True
    recur(0, 0)
```

#### Understanding the Role of the Boolean Return:

1. **Propagation of Success**: In the version with the boolean return statement, when a correct solution is found (i.e., the base case at `row == 9` is reached and the board is successfully filled), the function returns `True`. This `True` value is propagated back up through the recursive calls. It signals to each recursive level that the correct solution has been found, and there's no need to continue trying other numbers or paths. This prevents further unnecessary recursive calls after the solution is found.

2. **Triggering Backtracking**: When a call to `recur` does not lead to a solution (i.e., no valid number can be placed in the current cell or subsequent cells), the function returns `False` (either explicitly or implicitly by reaching the end of the function). This signals the need to backtrack: undo the last move and try the next possibility.

#### Without the Boolean Return:

Without using a return statement to propagate the success or failure of recursive calls:
- **Lack of Signal for Completion**: Your function lacks a mechanism to signal that a valid complete solution has been found. Consequently, even after finding a solution and printing the board, your code continues to execute. It will backtrack and attempt to find other solutions, unnecessarily modifying the board after the solution has been printed.
  
- **Ineffective Backtracking**: Without a return statement, the recursive function `recur(row, col + 1)` does not inform its caller when it has successfully completed or needs to backtrack. Thus, even if a solution is not possible along a path, your function doesn't effectively stop further exploration along that path, leading to incorrect or incomplete backtracking.

#### The Effect:

- The recursive calls continue even after finding and printing a solution, attempting to modify the board further, which is not the intended behavior in a Sudoku solver. The correct approach is to stop all recursion once the first solution is found (assuming you only need one solution), which is facilitated by the boolean return signaling mechanism.
